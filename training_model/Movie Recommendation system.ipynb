```python
# list string so we need to convert it first to list
import ast
ast.literal_eval(movies.iloc[1].genres)
```

```python
# [{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}]
# TO
# ['Adventure','Fantasy','Action']
def convert(columns):
    return [value['name'] for value in ast.literal_eval(columns)]
convert(movies['genres'][0])
```

```python
movies['genres'] = movies['genres'].map(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[1]['cast']
```

```python
def convert_cast(columns):
    cast_list = []
    i = 0
    for value in ast.literal_eval(columns):
        if i != 5:
            cast_list.append(value['name'])
            i+= 1
        else:
            break
    return cast_list
```

```python
movies['cast'].apply(convert_cast)[0]
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
movies.iloc[0]['crew']
```

```python
def convert_crew(columns):
    director_list = []
    i = 0
    for item in ast.literal_eval(columns):
        if item['job'] == "Director":
            director_list.append(item['name'])
    return director_list
```

```python
movies['crew'].map(convert_crew)
```

```python
movies['crew'] = movies['crew'].map(convert_crew)
```

```python
movies.head()
```

```python
movies['overview'].apply(lambda x: x.split(" "))
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split(" "))
```

```python
movies.head()
```

```python
movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies['overview'] = movies['overview'].apply(lambda x: [i.replace(" ","") for i in x])
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies = movies[['movie_id','title','tags']]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    stem_word = []
    for i in text.split():
        stem_word.append(ps.stem(i))
    return " ".join(stem_word)
```

```python
movies_dataset['tags'][0]
```

```python
stem(movies_dataset['tags'][0])
```

```python
movies_dataset['tags'][1]
```

```python
movies_dataset['tags'] = movies_dataset['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(tags_vector)
```

```python
similarity = cosine_similarity(tags_vector)
```

```python
def recommand(movie):
    movie_index = movies_dataset[movies_dataset['title'] == movie.lower()].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse= True, key=lambda x: x[1])[1:6]
    for i in movie_list:
        print(movies_dataset.iloc[i[0]].title)
```
